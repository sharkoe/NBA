```python
from nba_api.stats.static import teams

nba_teams = teams.get_teams()
# Select the dictionary for the Sixers, which contains their team ID
sixers = [team for team in nba_teams if team['abbreviation'] == 'PHI'][0]
sixers_id = sixers['id']
```

```python
sixers
```

```python
from nba_api.stats.endpoints import leaguegamefinder

# Query for games where the Celtics were playing
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=sixers_id)
# The first DataFrame of those returned is what we want.
games = gamefinder.get_data_frames()[0]
games.head()
pd.set_option('display.max_columns', None)
```

```python
season_2021 = games[games.SEASON_ID.str[-4:] == '2020']
```

```python
#only regular season games, dropped 2 preseason games
season_2021 = season_2021[:35]
```
